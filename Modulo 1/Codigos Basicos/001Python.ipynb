# Tutorial Variáveis


## O que são variáveis?

Variavéis são espaços na memória do computador onde podemos guardar informações (como textos, números, resultados de cálculos, etc) Para usar depois.

Variável com texto (string)

```python
nome = "Emanuel"
print(nome)
```

```python
nome = "waba"
```

```python
print(nome)
```

Variável com número inteiro (int)

```python
idade = 16
print(idade)
```

```python
# Este é um exemplo de número inteiro negativo
temperatura = -3 # Esta variável está armezenando um número negativo
print(temperatura) # print () é uma função de saída
```

Variável com número decimal (float)

```python
altura = 1.78 # Em python os números decimais são escritos com ".", exemplo= 5,2
print(altura)
```

```python
pi = 3.14
print(pi)
```

Variavél com valor boolenao (bool)

```python
estudante = True
print(estudante)
```

## O que é snake_case?

snake_case é um estilo de escrita usado para nome de variavéis, funções, e nomes em geral no código.
Neste estilo, as palavras são todas em minúsculo e separadas por um underlinde (_)

## Por que usar snake_case?

*   Deixa o código mais fácil de ler
*   É o padrão recomendado pea comunidade Python
*   Segue a PEF8, que é o guia de estilo ofial do Python

## OBS: PEF8 é um documento que recomenda boas práticas pra escrever código Python limpo e legível.
