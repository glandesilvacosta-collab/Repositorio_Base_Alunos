## Strings

É usado principalmente para armazebar informações de texto. Nelas temos uma sequência de caracteres que formam um texto. O Python permite formar strings com um par de aspas simples ou duplas

```python
"João"=="JOÃO" # um sinal de "=" significa atribuição. Dois sinais de "==" signfica comparação.
# O Python diferencia letras minusculas de maiusculas.
```

```python
nome = "André"
sobrenome = "Silva"
print(nome + " " + sobrenome) # Exemplo de concatenação. nome + " " (espaço) + sobrenome
```

# Operadores Aritméticos

### Adição (+)

```python
print(10 + 20)
print(10 + 20.5)
print(10 + 30 + 40)
```

```python
print(-89 + 63) # print() é uma função de saida no Python
print(23 - 17) # 23 + (-17)
print(23 + (-17)) # Se for fazer uma soma com dois sinais o sinal do número inteiro tem de ser isolado
print(34 - 67) # 34 + (-67)
```

### Subtração (-)

```python
print(123 - 87)
```

### Multiplicação

```python
print(10*2)
```

```python
print(2.4 * 5.2)
```

Divisão (/)

```python
print(1000/10)
print(129/23)
```

Divisão com resultado inteiro (//)

```python
print(129//23)
```

### Exponenciação

```python
print(2**3) # 2³ = 2.2.2 = 8
```

```python
print(2**-1) # Serve também para expoentes em negativo
```

```python
print(2**-1.5)
```

### Precedência

1.   Parênteses
2.   Expoentes
3.   Multiplicações e Divisão (Da esquerda para a direita)
5.   Somas e subtração


```python
5 + 25 * 2
```

```python
(5 + 25) * 2 # Como o parênteses está "maior" na hierarquia ele será lido primeiro
```

```python
(12 + 3 - 1 * 50/3)
```

```python
4 ** 0.5 # Tirando a raiz quadrada de 2
```

# Hora da Prática

1) Crie um código que receba o nome do aluno, duas notas e informe a média do aluno.
OBS: No seu print você deve apresentar o nome do aluno, as duas notas e a média

```python
nomedoaluno = input("Digite o nome do aluno: ")
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
media = (nota1 + nota2) / 2

print(f"O aluno(a) {nomedoaluno} teve notas {nota1} e {nota2}, e sua média é {media}")
```

2) Crie um código que receba as medias de um retângulo e informe o perimetro do mesmo.

```python
comprimento = float(input("Digite o comprimento do retângulo: "))
largura = float(input("Digite a largura do retângulo: "))
perimetroretangulo = 2 * (comprimento + largura)

print(f"O perímetro do retângulo é: {perimetroretangulo}")
```

3) Crie um código que calcule a raiz quadrada de um número

```python
numero = float(input("Digite um número para calcular a raiz quadrada: "))
raizquadrada = numero ** 0.5
print(f"A raiz quadrada de {numero} é {raizquadrada}")
```

4) Crie uma calculadora simples (adição, subtração, multiplicação e divisão). Seu código deve receber 2 números e apresentar o resultado das 4 operações

```python
num1 = float(input("Digite um número: "))
num2 = float(input("Digite outro número: "))

print(f"A soma de {num1} e {num2} é {num1 + num2}")
print(f"A subtração de {num1} e {num2} é {num1 - num2}")
print(f"A multiplicação de {num1} e {num2} é {num1 * num2}")
print(f"A divisão de {num1} e {num2} é {num1 / num2}")
```

5) Calcule a diferença entre dois números

```python
num1 = int(input("Digite um número: "))
num2 = int(input("Digite outro número: "))

print(f"A diferença entre {num1} e {num2} é {num1 - num2}")
```

6) Calcule o produto entre dois números

```python
num1 = float(input("Digite um número: "))
num2 = float(input("Digite outro número: "))

print(f"A multiplicação de {num1} e {num2} é {num1 * num2}")
```

7) Crie um código que receba uma base e depois o expoente e calcule a potencia

```python
num1 = float(input("Digite a base: "))
num2 = float(input("Digite o expoente: "))

print(f"O Exponente de {num1} é {num1 ** num2}")
```

8) Crie um programa que solicite ao usuário a quantidade de horas trabalhadas por semana e o valor da hora de trabalho e calcule o salário mensal

```python
num1 = float(input("Digite a quantidade de horas trabalhadas por semana: "))
num2 = float(input("Digite o valor da hora: "))

salariomensal = num1 * 4.2 * num2

print(f"O seu salário mensal é de {salariomensal}")
```
